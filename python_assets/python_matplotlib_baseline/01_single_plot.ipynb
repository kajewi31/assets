```python
import numpy as np

import matplotlib.pyplot as plt

x = np.linspace(1, 10, 100)
sin_y = np.sin(x)
```

```python
# A
plt.figure(figsize=(5, 5))
plt.plot(x, sin_y)
plt.tight_layout()
plt.show()
```

```python
# B
fig = plt.figure(figsize=(5, 5))
# ax = fig.add_subplot(1, 1, 1)
ax = fig.add_subplot()
ax.plot(x, sin_y)

plt.tight_layout()
plt.show()
```

```python
# C
fig, ax = plt.subplots(1, 1, figsize=(5, 5))
ax.plot(x, sin_y)
plt.tight_layout()
plt.show()
```

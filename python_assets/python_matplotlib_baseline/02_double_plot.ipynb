```python
import numpy as np

import matplotlib.pyplot as plt

x = np.linspace(1, 10, 100)
sin_y = np.sin(x)
cos_y = np.cos(x)
```

```python
# A
plt.figure(figsize=(10, 5))
plt.subplot(2, 1, 1)
plt.plot(x, sin_y)

plt.subplot(2, 1, 2)
plt.plot(x, cos_y)
plt.tight_layout()
plt.show()
```

```python
# B
fig = plt.figure(figsize=(10, 5))
ax_1 = fig.add_subplot(2, 1, 1)
ax_1.plot(x, sin_y)


ax_2 = fig.add_subplot(2, 1, 2)
ax_2.plot(x, cos_y)
plt.tight_layout()
plt.show()
```

```python
# C
fig, axs = plt.subplots(2, 1, figsize=(10, 5))
axs[0].plot(x, sin_y)
axs[1].plot(x, cos_y)

plt.tight_layout()
plt.show()
```

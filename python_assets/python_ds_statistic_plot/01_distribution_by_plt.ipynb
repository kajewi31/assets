```python
import pandas as pd
import numpy as np
```

```python
left_long_tail_samples = -((np.random.pareto(a=10, size=1000) + 1))
right_long_tail_samples = (np.random.pareto(a=10, size=1000) + 1)
normal_samples_samples = np.random.normal(loc=0, scale=1, size=1000)
```

```python
import matplotlib.pyplot as plt
```

```python
# plot
fig, axes = plt.subplots(3, 1, sharex=True,figsize=(10,5))

ax = axes[0]
ax.hist(normal_samples_samples)
ax.set_title('Histogram')

ax = axes[1]
ax.boxplot(normal_samples_samples, vert=False)
ax.set_title('BoxPlot')

ax = axes[2]
ax.violinplot(normal_samples_samples, vert=False)
ax.set_title('ViolinPlot')

plt.tight_layout()
plt.show()
```

```python
# plot
fig, axes = plt.subplots(3, 1, sharex=True,figsize=(10,5))

ax = axes[0]
ax.hist(left_long_tail_samples)
ax.set_title('Histogram')

ax = axes[1]
ax.boxplot(left_long_tail_samples, vert=False)
ax.set_title('BoxPlot')

ax = axes[2]
ax.violinplot(left_long_tail_samples, vert=False)
ax.set_title('ViolinPlot')

plt.tight_layout()
plt.show()
```

```python
import numpy as np

x = np.linspace(1, 10, 100)
sin_y = np.sin(x)
cos_y = np.cos(x)
```

```python
# B
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(
    go.Scatter(x=x, y=sin_y, mode='lines', name='Sine Wave')
)
fig.add_trace(
    go.Scatter(x=x, y=cos_y, mode='lines', name='Cosine Wave')
)

fig.update_layout(
    title="Sine and Cosine Waves",
    xaxis_title="X-axis", yaxis_title="Y-axis",
    width=600, height=400
)
fig.show()
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

#  
fig = make_subplots(rows=1, cols=1, shared_xaxes=True,
                    subplot_titles=("Sine Wave", "Cosine Wave"))

#    (Sine)
fig.add_trace(go.Scatter(x=x, y=sin_y, mode='lines', name='Sine Wave'), row=1, col=1)

#    (Cosine)
fig.add_trace(go.Scatter(x=x, y=cos_y, mode='lines', name='Cosine Wave'), row=1, col=1)

fig.update_layout(
    title="Sine and Cosine Waves with Subplots",
    height=600, width=600,
    xaxis_title="X-axis", yaxis_title="Y-axis"
)
fig.show()
```

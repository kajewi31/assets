```python
import numpy as np

x = np.linspace(1, 10, 100)
sin_y = np.sin(x)
```

```python
# A
import plotly.express as px

fig = px.line(x=x, y=sin_y,
              title="Sine Wave", labels={'x': 'X-axis', 'y': 'Sine'},
              width=600, height=400
              )
fig.show()
```

```python
# B
import plotly.graph_objects as go


fig = go.Figure()
fig.add_trace(
    go.Scatter(x=x, y=sin_y, mode='lines', name='Sine Wave')
)

fig.update_layout(
    title="Sine Wave",
    xaxis_title="X-axis", yaxis_title="Sine",
    width=600, height=400
)
fig.show()
```

```python
# C
from plotly.subplots import make_subplots

fig = make_subplots(rows=1, cols=1)

fig.add_trace(go.Scatter(x=x, y=sin_y, mode='lines', name='Sine Wave'), row=1, col=1)

fig.update_layout(
    title="Sine Wave with Subplots",
    xaxis_title="X-axis", yaxis_title="Sine",
    width=600, height=400

)
fig.show()
```

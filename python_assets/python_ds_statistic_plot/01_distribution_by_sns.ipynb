```python
import pandas as pd
import numpy as np
```

```python
left_long_tail_samples = -((np.random.pareto(a=10, size=1000) + 1))
right_long_tail_samples = (np.random.pareto(a=10, size=1000) + 1)
normal_samples_samples = np.random.normal(loc=0, scale=1, size=1000)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# plot
fig, axes = plt.subplots(3, 1, sharex=True, figsize=(10, 5))

ax = axes[0]
sns.histplot(normal_samples_samples, ax=ax, element='bars')
ax.set_title('sns.histplot - element:bars (default)')

ax = axes[1]
sns.histplot(normal_samples_samples, ax=ax, element='step')
ax.set_title('sns.histplot - element:step')

ax = axes[2]
sns.histplot(normal_samples_samples, ax=ax, element='poly')
ax.set_title('sns.histplot - element:poly')

plt.tight_layout()
plt.show()
```

```python
# plot
fig, axes = plt.subplots(4, 1, sharex=True, figsize=(10, 5))

ax = axes[0]
sns.kdeplot(normal_samples_samples, ax=ax)
ax.set_title('[sns.kdeplot - RAW]')

ax = axes[1]
sns.histplot(normal_samples_samples, ax=ax, stat='density', element='step', alpha=0.3)
sns.kdeplot(normal_samples_samples, ax=ax)
ax.set_title('[sns.kdeplot - RAW] with [sns.histplot]')

ax = axes[2]
sns.kdeplot(normal_samples_samples, ax=ax, fill=True)
ax.set_title('[sns.kdeplot - fill=True]')

ax = axes[3]
sns.histplot(normal_samples_samples, ax=ax, stat='density', element='step', fill=False)
sns.kdeplot(normal_samples_samples, ax=ax, fill=True)
ax.set_title('[sns.kdeplot - fill=True] with [sns.histplot]')

plt.tight_layout()
plt.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt

x = np.linspace(1, 10, 100)
sin_y = np.sin(x)
cos_y = np.cos(x)
tan_y = np.tan(x)
```

```python
# A
plt.figure(figsize=(10, 5))
plt.subplot(3, 1, 1)
plt.plot(x, sin_y)

plt.subplot(3, 1, 2)
plt.plot(x, cos_y)

plt.subplot(3, 1, 3)
plt.plot(x, tan_y)
plt.tight_layout()
plt.show()
```

```python
# B
fig = plt.figure(figsize=(10, 5))
ax_1 = fig.add_subplot(3, 1, 1)
ax_1.plot(x, sin_y)

ax_2 = fig.add_subplot(3, 1, 2)
ax_2.plot(x, cos_y)

ax_3 = fig.add_subplot(3, 1, 3)
ax_3.plot(x, tan_y)

plt.tight_layout()
plt.show()
```

```python
# C
fig, axs = plt.subplots(3, 1, figsize=(10, 5))
axs[0].plot(x, sin_y)
axs[1].plot(x, cos_y)
axs[2].plot(x, tan_y)

plt.tight_layout()
plt.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt

x = np.linspace(1, 10, 100)

sin_y_a = np.sin(x)
cos_y_a = np.cos(x)

sin_y_b = np.sin(x * 2)
cos_y_b = np.cos(x * 2)
```

```python
# A
plt.figure(figsize=(10, 5))
plt.subplot(2, 2, 1)
plt.plot(x, sin_y_a)
plt.title('sin_y_a')

plt.subplot(2, 2, 2)
plt.plot(x, cos_y_a)
plt.title('cos_y_a')

plt.subplot(2, 2, 3)
plt.plot(x, sin_y_b)
plt.title('sin_y_b')

plt.subplot(2, 2, 4)
plt.plot(x, cos_y_b)
plt.title('cos_y_b')
plt.tight_layout()
plt.show()
```

```python
# B
fig = plt.figure(figsize=(10, 5))
ax_1_1 = fig.add_subplot(2, 2, 1)
ax_1_1.plot(x, sin_y_a)
ax_1_1.set_title('sin_y_a')

ax_1_2 = fig.add_subplot(2, 2, 2)
ax_1_2.plot(x, sin_y_a)
ax_1_2.set_title('cos_y_a')

ax_2_1 = fig.add_subplot(2, 2, 3)
ax_2_1.plot(x, sin_y_a)
ax_2_1.set_title('sin_y_b')

ax_2_2 = fig.add_subplot(2, 2, 4)
ax_2_2.plot(x, sin_y_a)
ax_2_2.set_title('cos_y_b')

plt.tight_layout()
plt.show()
```

```python
# C
fig, axs = plt.subplots(2, 2, figsize=(10, 5))

axs[0][0].plot(x, sin_y_a)
axs[0][0].set_title('sin_y_a')

axs[0][1].plot(x, cos_y_a)
axs[0][1].set_title('cos_y_a')

axs[1][0].plot(x, sin_y_b)
axs[1][0].set_title('sin_y_b')

axs[1][1].plot(x, cos_y_b)
axs[1][1].set_title('cos_y_b')

plt.tight_layout()
plt.show()
```

```python
# C
fig, axs = plt.subplots(2, 2, figsize=(10, 5))
axs = axs.flatten()

axs[0].plot(x, sin_y_a)
axs[0].set_title('sin_y_a')

axs[1].plot(x, cos_y_a)
axs[1].set_title('cos_y_a')

axs[2].plot(x, sin_y_b)
axs[2].set_title('sin_y_b')

axs[3].plot(x, cos_y_b)
axs[3].set_title('cos_y_b')

plt.tight_layout()
plt.show()
```

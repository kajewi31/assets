```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
```

```python
data_left_skewed = np.random.beta(a=1, b=5, size=1000) * 100  # Beta distribution for left-skewed data
df_left = pd.DataFrame({"value": data_left_skewed})
```

```python
# draw raw distribution
sns.kdeplot(df_left['value'], fill=True)
plt.axvline(df_left['value'].mean(), color='orange', linestyle='-', linewidth=2, label='Mean')
plt.axvline(df_left['value'].median(), color='purple', linestyle='-', linewidth=2, label='Median')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# raw train_test_split distribution
trains, tests = train_test_split(df_left['value'], test_size=0.2, random_state=42)

fig, axes = plt.subplots(3, 1, figsize=(8, 5), sharex=True)

axes[0].boxplot(df_left['value'], vert=False)
axes[0].set_title("Box Plot of Raw Data")

axes[1].boxplot(trains, vert=False)
axes[1].set_title("Box Plot of Train Data")

axes[2].boxplot(tests, vert=False)
axes[2].set_title("Box Plot of Test Data")

plt.tight_layout()
plt.show()
```

```python
# cut_stratified train_test_split distribution

cut_series = pd.cut(df_left['value'], 7, labels=False)
trains, tests, _, _ = train_test_split(df_left['value'], df_left['value'], test_size=0.2, random_state=42, stratify=cut_series)

fig, axes = plt.subplots(3, 1, figsize=(8, 5), sharex=True)

axes[0].boxplot(df_left['value'], vert=False)
axes[0].set_title("Box Plot of Raw Data")

axes[1].boxplot(trains, vert=False)
axes[1].set_title("Box Plot of Train Data")

axes[2].boxplot(tests, vert=False)
axes[2].set_title("Box Plot of Test Data")
fig.suptitle('Cut stratifying')
plt.tight_layout()
plt.show()
```

```python
# qcut_stratified train_test_split distribution

qcut_series = pd.qcut(df_left['value'], 7, labels=False)
trains, tests, _, _ = train_test_split(df_left['value'], df_left['value'], test_size=0.2, random_state=42, stratify=qcut_series)

fig, axes = plt.subplots(3, 1, figsize=(8, 5), sharex=True)

axes[0].boxplot(df_left['value'], vert=False)
axes[0].set_title("Box Plot of Raw Data")

axes[1].boxplot(trains, vert=False)
axes[1].set_title("Box Plot of Train Data")

axes[2].boxplot(tests, vert=False)
axes[2].set_title("Box Plot of Test Data")
fig.suptitle('Qcut stratifying')
plt.tight_layout()
plt.show()
```
